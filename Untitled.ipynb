```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, SimpleRNN, GRU

# Load your data
data = pd.read_csv('data.csv')  # Replace with your file
data['date'] = pd.to_datetime(data['date'])
data = data[['date', 'warehouse_capacity', 'truck_capacity']]
data.set_index('date', inplace=True)

# Normalize data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)
```

```python
# Prepare sequences for LSTM
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i+sequence_length])
        y.append(data[i+sequence_length])
    return np.array(X), np.array(y)

sequence_length = 5  # Look back 30 days
X, y = create_sequences(scaled_data, sequence_length)

```

```python
# Split into train and test sets
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

```

```python
# Build LSTM model
def lstm_model():
    model = Sequential([
        LSTM(50, activation='relu', return_sequences = True, input_shape=(X_train.shape[1], X_train.shape[2])),
        Dropout(0.1),
        LSTM(200, activation='relu'),
        Dropout(0.1),
        Dense(y_train.shape[1])  # Output layer
    ])

    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
# Build RNN model
def rnn_model():
    model = Sequential([
        SimpleRNN(50, activation='relu', return_sequences = True, input_shape=(X_train.shape[1], X_train.shape[2])),
        Dropout(0.1),
        SimpleRNN(200, activation='relu'),
        Dropout(0.1),
        Dense(y_train.shape[1])  # Output layer
    ])

    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
# Build GRU model
def gru_model():
    model = Sequential([
        GRU(50, activation='relu', return_sequences = True, input_shape=(X_train.shape[1], X_train.shape[2])),
        Dropout(0.1),
        GRU(200, activation='relu'),
        Dropout(0.1),
        Dense(y_train.shape[1])  # Output layer
    ])

    model.compile(optimizer='adam', loss='mse')
    return model
```

```python
model_lstm = lstm_model()
history = model_lstm.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
def predict(model):
    # Predict the next 14 days
    last_sequence = scaled_data[-sequence_length:]
    predicted = []
    
    for _ in range(14):
        prediction = model.predict(last_sequence[np.newaxis, :, :])
        predicted.append(prediction[0])
        last_sequence = np.vstack((last_sequence[1:], prediction))
    
    predicted = scaler.inverse_transform(predicted)


    # Display predictions
    predicted_df = pd.DataFrame(predicted, columns=data.columns)
    predicted_df.index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=14)
    predicted_df['warehouse_capacity'] = predicted_df['warehouse_capacity'].astype(int)
    predicted_df['truck_capacity'] = predicted_df['truck_capacity'].astype(int)
    # print(predicted_df)
    return predicted_df
```
